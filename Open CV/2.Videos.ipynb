```python
frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
print(frames)
```

```python
fr_sec = cap.get(cv2.CAP_PROP_FPS)
print(fr_sec)
```

Working with the Video

```python
cap = cv2.VideoCapture('Data/1.mp4')
# Loop through the frames
for frame_idx in range(int(frames)):
    ret, frame = cap.read() # Read the next frame
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Convert to grayscale
    cv2.imshow('frame', gray) # Show the frame
    if cv2.waitKey(10) & 0xFF == ord('q'): # Hit `q` to exit
        break


# Release the VideoCapture object
cap.release()
cv2.destroyAllWindows()
```

Write out a Video

```python
cap = cv2.VideoCapture('Data/1.mp4')
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
fps = cap.get(cv2.CAP_PROP_FPS)
# Video Writer
fourcc = cv2.VideoWriter_fourcc("P", "I", "M", "1")
out = cv2.VideoWriter('Data/output.avi', fourcc, fps, (int(width), int(height)), isColor=False)
# Loop through the frames
for frame_idx in range(int(frames)):
    ret, frame = cap.read() # Read the next frame

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY) # Convert to grayscale
    
    cv2.imshow('frame', gray) # Show the frame
    out.write(gray) # Write the frame

    if cv2.waitKey(10) & 0xFF == ord('q'): # Hit `q` to exit
        break


# Release the VideoCapture object
cap.release()
cv2.destroyAllWindows()
out.release()
```
