```python
new_hdf = pd.get_dummies(h_df, drop_first=True)
```

```python
X = new_hdf.drop('HeartDisease', axis=1)
y = new_hdf['HeartDisease']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
scores = cross_val_score(SVC(), X_train, y_train, cv=5)
scores.mean()
```

```python
bag = BaggingClassifier(SVC(), max_samples=0.8, n_estimators=100, random_state=42)
scores = cross_val_score(bag, X_train, y_train, cv=5)
scores.mean()
```

```python
scores = cross_val_score(RandomForestClassifier(), X_train, y_train, cv=5)
scores.mean()
```
